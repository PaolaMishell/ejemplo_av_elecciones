```python
# Si quieres una muestra con un porcentaje específico, por ejemplo, 10%:
df_randomica = df_limpio.sample(frac=0.1, random_state=42)
```

```python
# Guardar el DataFrame con la nueva columna
df_randomica.to_excel("/content/salert-randomica.xlsx", index=False)
```

```python
# Función para clasificar el voto
def clasificar_voto(texto):
    texto = texto.lower()
    if "noboa" in texto:
        return "Voto Noboa"
    elif "luisa" in texto or "gonzález" in texto:
        return "Voto Luisa"
    else:
        return "Nulo"
```

```python
# Aplicar la función a la columna 'text' y crear una nueva columna 'voto'
df_randomica["etiqueta"] = df_randomica["text"].apply(clasificar_voto)
```

```python
# Mostrar el resultado
df_randomica
```

```python
# Guardar el DataFrame con la nueva columna
df_randomica.to_excel("/content/salert-randomica-clasificado.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Contar la cantidad de votos por categoría
conteo_votos = df_randomica["etiqueta"].value_counts()

# Configurar el estilo de la gráfica
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))

# Crear gráfico de barras
sns.barplot(x=conteo_votos.index, y=conteo_votos.values, palette=["blue", "red", "gray"])

# Etiquetas y título
plt.xlabel("Categoría de Voto", fontsize=12)
plt.ylabel("Cantidad", fontsize=12)
plt.title("Distribución de Votos en la Muestra", fontsize=14)

# Mostrar valores encima de las barras
for i, v in enumerate(conteo_votos.values):
    plt.text(i, v + 5, str(v), ha="center", fontsize=12)

# Mostrar la gráfica
plt.show()
```
